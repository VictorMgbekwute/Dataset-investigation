## Introduction

### Data Description

This Dataset gives the information about 100k different patients that either missed or showed up during their appointment in Brazil. The dataset consist of 110527 rows and 14 columns which include: PatientId, Gender, ScheduledDay, AppointmentDay, AppointmentID, Handcap, Age, Neighbourhood, Scholarship, Hipertension, Diabetes, Alcoholism, SMS_received, No-show

### Questions for Analysis
Question1: How often do people with scholarship attend appointment as compare to others?
Question2: How will the type of illness affect attending an appointment.

```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Analysis

### Research Question 1 (How SMS_received affect attending an appointment )

```python
# This gives us the patient that did not attend their appointment after receiving SMS
no_show= project_df.query("No_show == 'Yes'")
```

```python
no_show.head(8)
```

```python
def countplot_function(x,title): 
    sns.countplot(x,data = no_show, hue = x).set(Title =title)
```

```python
# Variation between those that attended the appointment and those that did not.
countplot_function(x='SMS_received','Plot of appointment attendance through SMS_received')
```

### Research Question 2  (Did patients with Scholarship attend more Appointment!)

```python
# Graph showing how scholarship affect appointment time
SC = sns.countplot(x="Scholarship",  hue="Scholarship", data= no_show)
```

```python
# To get the number of those that received SMS and those that did not
AS_count= no_show['SMS_received'].value_counts()
AS_count
```

```python
# To get the number of those that with Scholarship and those without Scholarship
SC_count= no_show['Scholarship'].value_counts()
SC_count
```

```python
# This graph shows the relationship between that that have Scholarship and did not atten their appointment and those that received SMS and still did not attend their appointment.
plt.scatter(x=AS_count,y=SC_count);
```

```python
### Did the type of illness affect appointment time
```

```python
# To get the number of those with Hipertension and those without
Hi_count= no_show['Hipertension'].value_counts()
Hi_count
```

```python
# A graph that shows the relationship between those that did not attend their appointment and those that did.
no_show['Hipertension'].plot(kind='hist')
```

```python
# To get the number of those with Diabetes and those without
Di_count= no_show['Diabetes'].value_counts()
Di_count
```

```python
# A graph that shows the relationship between those with Diabetes that did not attend their appointment and those that did.
no_show['Diabetes'].plot(kind='hist')
```

```python
# To get the number of those that drink Alcohol and those that did not
Al_count= no_show['Alcoholism'].value_counts()
Al_count
```

```python
# A graph that shows the relationship between those that drink Alcohol and did not attend their appointment and those that attended that appointment.
no_show['Alcoholism'].plot(kind='hist')
```

## Conclusions
From my first graph, I compared if SMS_received will help in keeping an appointment and it shows that even does that did not received any SMS kept their appointment more often than does that received at least one SMS.
Further analysis with Scholarship shows that despite been enrolled in Brasilian Welfare many patients did not keep to their appointment time. Also there is a higher correlation between Scholarship and SMS received. 
### Limitation
The reason for the Patients not keeping to appointment might results from the fact that maybe the SMS was not sent on time or the distance form the Neighbourhood to the hospital is far.
